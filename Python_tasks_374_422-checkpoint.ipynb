Напишите функцию, которая считает факториал числа с помощью рекурсии.

Кстати, в Python есть ограничение на количество рекурсивных вызовов. Попробуйте передать своей функции, например, число 1000 и посмотрите, что будет.

```python
def factorial(num):
    
    if num == 1:
        return 1
    fact_n_minus_1 = factorial(num - 1)
    return num * fact_n_minus_1

num_fact = factorial(5)
print(num_fact)
```

def power(a, n):

    return a * power(a, n)


float_num = float(input('Введите вещественное число: '))

Другие пользователи отметили, что это решение нерабочее и в нём есть ошибки. Исправьте это решение, не используя циклы, возведение в степень через ** и функцию math.pow()

Правильный результат:

Введите вещественное число: 1.5

#### 376 - Поиск элемента

Когда мы работаем с большой многоуровневой структурой, нам нередко необходимо пройтись по ней и найти нужный элемент. Для этого в программировании используются специальные алгоритмы поиска.

Напишите функцию, которая находит заданный пользователем ключ в словаре и выдаёт значение этого ключа на экран. В качестве примера можно использовать такой словарь.

Пример 1:

Искомый ключ: h2

Значение: Здесь будет мой заголовок


Пример 2:

Искомый ключ: abc

Такого ключа в структуре сайта нет.

```python
site = {
    'html': {
        'head': {
            'title': 'Мой сайт'
        },
        'body': {
            'h2': 'Здесь будет мой заголовок',
            'div': 'Тут, наверное, какой-то блок',
            'p': 'А вот здесь новый абзац'
        }
    }
}

def find_key(struct, key):
    
    if key in struct:
        return struct[key]
    
    for sub_struct in struct.values():
        if isinstance(sub_struct, dict):
            result = find_key(sub_struct, key)
            if result:
                break
    else:
        result = None
        
    return result

user_key = input('Искомый ключ: ')
value = find_key(site, user_key)
if value:
    print('Значение:', value)
else:
    print('Такого ключа в структуре сайта нет.')
```

Вопрос — обязательный позиционный аргумент, остальные — со значениями по умолчанию. При корректном ответе функция может возвращать что угодно — например, число 1 при ответе «да» или 0 при ответе «нет».

В основной программе вызовите функцию минимум три раза: только с вопросом, с вопросом и сообщением об ошибке, с вопросом и количеством попыток.

Пример работы программы:

Вы действительно хотите выйти? что

```python
def ask_user(question,                                                     # вопрос
             complaint ='Неверный ввод. Пожалуйста, введите да или нет',   # ошибка
             retries=4):                                                   # кол-во попыток
    
    while True:

        answer = input(question).lower()

        if answer == 'да':
            return 1
        if answer == 'нет':
            return 0
        retries -= 1
        if retries == 0:
            print('Кол-во попыток истекло.')
            break

        print(complaint)
        print('Осталось попыток:', retries)
        
ask_user('Вы действительно хотите выйти?')        
ask_user('Удалить файл?', 'Так удалить или нет?')
ask_user('Записать файл?', retries=2)
```

Пользователь вводит искомый ключ. Если он хочет, то может ввести максимальную глубину — уровень, до которого будет просматриваться структура. Напишите функцию, которая находит заданный пользователем ключ в словаре и выдаёт значение этого ключа на экран. По умолчанию уровень не задан. В качестве примера можно использовать такой словарь.

```python
site = {
    'html': {
        'head': {
            'title': 'Мой сайт'
        },
        'body': {
            'h2': 'Здесь будет мой заголовок',
            'div': 'Тут, наверное, какой-то блок',
            'p': 'А вот здесь новый абзац'
        }
    }
}

def find_key(data, key, depth=-1):
    
    if depth == 0:
        return None
        
    for k, v in data.items():
        if k == key:
            return v
        else:
            if isinstance(v, dict):
                res = find_key(v, key, depth - 1)
                if res != None:
                    return res
                
    return None

key = input('Введите ключ: ')
depth = int(input('Введите глубину: '))
find_key(site, key, depth)
```

Напишите свою функцию sum, которая должна быть более гибкой, чем стандартная функция sum. Вот что она должна уметь делать:

Складывать числа из списка списков.

Складывать из набора параметров.

Примеры вызовов функции:

sum([[1, 2, [3]], [1], 3])

Ответ: 10

sum(1, 2, 3, 4, 5)

Ответ: 15

```python
def my_sum(*data, ar=None):
    
    if not ar:
        ar = []
        
    for elem in data:
        if isinstance(elem, list):
            ar.append(my_sum(*elem))
        else:
            ar.append(elem)
            
    return sum(ar)
            
print('Ответ:', my_sum(1, 2, 3, 4, 5))
```

```python
def caesar_cipher(string, user_shift):
    
    char_list = [(alphabet[(alphabet.index(sym.upper()) + user_shift) % 33] if sym != ' ' else ' ') for sym in string]
    new_str = ''.join(char_list)
    return new_str

alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

file = open(r'task\text.txt', 'r')
result = []
a = 0
for i in file:
    a += 1
    result.append(caesar_cipher(i[0: len(i) - 1], a))
file.close()
file = open(r'task\cipher_text.txt', 'w')
for elem in result:
    file.write(elem + '\n')
file.close()
```

Дан файл words.txt, в котором построчно записаны слова. Необходимо определить количество слов, из которых можно получить палиндром (привет предыдущим модулям). Если в строке встречается число, то программа выдаёт ошибку ValueError и записывает эту ошибку в файл errors.log

```python
def is_poly(string):      # определяет палиндром
    
    char_dict = {}
    for i_sym in string:
        char_dict[i_sym] = char_dict.get(i_sym, 0) + 1  # идем по символам строки и если такого символа нет, то возвращаем 0
                                                        # а после этого прибавляем 1, т к мы его только что встретили
    odd_count = 0
    for i_value in char_dict.values():
        if i_value % 2 != 0:
            odd_count += 1
            
    return odd_count <= 1

try:
    file = open(r'task\words.txt', 'r', encoding="utf-8")
    words_pal = 0
    for i_line in file:
        for elem in i_line.rstrip():
            if not elem.isalpha():
                raise ValueError()
        if is_poly(i_line.rstrip()):
            words_pal += 1
    
except ValueError:
    print('В строке встретилось число!')
    file = open(r'task\errors.log', 'w')
    file.write('В строке встретилось число!')
finally:
    print('Количество слов, из которых можно получить палиндром:', words_pal)
file.close()
```

#### 415 - Сумма символов в строке 3

Рассмотрим предыдущую задачу, где считается сумма символов. А если в какой-то строке файла меньше трех символов, то мы выдаем на экран ошибку:

```python
# В этой программе, кстати, можно заметить еще одну не очень красивую штуку, а именно строчку people_file.close()
# Конечно, при вызове исключения программа просто упадет и работа с файлом закончится,
# Но лучше специально задать себе вопрос: а вдруг не закроется?
# И вдруг мы будем как-то еще обрабатывать эту ошибку и файл нужен будет где-то еще?
# К тому же, такой код много кому будет резать глаза

# Чтобы глаза и программа были в безопасности, лучше использовать специальный оператор - with
# Оператор with правильно называть "контекстным менеджером"

# Конструкция with - as используется для безопасной работы с файлом
# Это значит, что как только этот блок кода завершается, и возникает исключение, 
# то при любом выходе из этого кода, оператор with закроет этот файл сам
# Поэтому теперь можно убрать лишние строки с открытием и закрытием файла:

sym_sum = 0
line_count = 0
try:
    with open(r'task\people.txt', 'r', encoding="utf-8") as people_file:    # as - вместо оператора присваивания
        for i_line in people_file:
            lenght = len(i_line)
            line_count += 1
            if i_line.endswith('\n'):  
                lenght -= 1
            if lenght < 3:
                raise BaseException('Длина {} строки меньше трех символов.'.format(line_count))
            sym_sum += lenght
    
except FileNotFoundError:
    print('Файл не найден.')
finally:
    print('Найденная сумма символов:', sym_sum) 
    print(people_file.closed)
```

Например:

Василий test@test.ru 27


Напишите программу, которая проверяет данные из файла для каждой строки:

Присутствуют все три поля.

Поле имени содержит только буквы.

Поле «Имейл» содержит @ и . (точку).

Поле «Возраст» является числом от 10 до 99.

В результате проверки сформируйте два файла:

registrations_good.log — для правильных данных, записывать строки как есть,

registrations_bad.log — для ошибочных, записывать строку и вид ошибки.

Для валидации строки данных напишите функцию, которая может выдавать исключения:

НЕ присутствуют все три поля: ValueError.

Поле имени содержит НЕ только буквы: NameError.

Поле «Имейл» НЕ содержит @ и .(точку): SyntaxError.

Поле «Возраст» НЕ является числом от 10 до 99: ValueError. 

```python
# Вариант 1:

def func_errors():
    
    with open(r'task\registrations.txt', 'r', encoding='utf-8') as file:
        for line in file:
            try:
                good = True
                tmp = line.split()
                if not len(tmp) == 3:
                    raise ValueError()
                    good = False
                else:
                    error = False
                    for a in tmp[0]:
                        if not a.isalpha():
                            error = True
                    if error:
                        raise NameError()
                        good = False
                    if tmp[1].find('@') == -1 or tmp[1].find('.') == -1:
                        raise SyntaxError()
                        good = False
                    if int(tmp[2]) < 10 or int(tmp[2]) > 99:
                        raise ValueError()
                        good = False
                with open(r'task\registrations_good.log', 'a') as file_2:
                    file_2.write(line)

            except ValueError:
                print('НЕ присутствуют все три поля')
                with open(r'task\registrations_bad.log', 'a') as file_3:
                        file_3.write(line)
            except NameError:
                print('Поле имени содержит НЕ только буквы')
                with open(r'task\registrations_bad.log', 'a') as file_3:
                        file_3.write(line)
            except SyntaxError:
                print('Поле «Имейл» НЕ содержит @ и .(точку)')
                with open(r'task\registrations_bad.log', 'a') as file_3:
                        file_3.write(line)
            except ValueError:
                print('Поле «Возраст» НЕ является числом от 10 до 99')
                with open(r'task\registrations_bad.log', 'a') as file_3:
                        file_3.write(line)
func_errors()
```

```python
# Вариант 2:

def validator(string):  # функция для проверки ошибок
    list = string.split()
    if len(list) < 3:
        raise IndexError('НЕ присутствуют все три поля')
    elif not list[0].isalpha():
        raise NameError('Поле имени содержит НЕ только буквы')
    elif not int(list[2]) in range(10, 100):
        raise ValueError('Поле «Возраст» НЕ является числом от 10 до 99')
    elif (list[1].count('@') + list[1].count('.')) != 2:
        raise SyntaxError('Поле «Имейл» НЕ содержит @ и .(точку)')
        
def log(logname, message):  # функция для записи
    with open(logname, 'a') as logfile:
        logfile.write(message + '\n')
        
access_log = 'registrations_good.log'
error_log = 'registrations_bad.log'

with open(r'task\registrations.txt') as file:
    for line in file.readlines():
        try:
            validator(line.strip())
        except IndexError as error:
            log(error_log, f'{line.strip()} {repr(error)}')
        except NameError as error:
            log(error_log, f'{line.strip()} {repr(error)}')
        except ValueError as error:
            log(error_log, f'{line.strip()} {repr(error)}')
        except SyntaxError as error:
            log(error_log, f'{line.strip()} {repr(error)}')
        else:
            log(access_log, line.strip())
```

#### 420 - Текстовый калькулятор

Иван стоит на пороге величайшего открытия (не будем его расстраивать), которое перевернёт представление о всей математике и программировании. Имя этому открытию — текстовый калькулятор. Правда, код для своего открытия ему сложно написать самому, и поэтому он попросил вас помочь ему. Так что уже можно бежать в патентное бюро.

Есть файл calc.txt, в котором хранятся записи вида

100 + 34

23 / 4

то есть ОПЕРАНД_1 ОПЕРАЦИЯ ОПЕРАНД_2, разделённые пробелами.

Операнды — целые числа. Операции — арифметические (включая деление нацело и нахождение остатка).


Напишите программу, которая вычисляет все эти операции и находит сумму их результатов. Пропишите обработку возможных ошибок. Программа не должна завершаться при первой же ошибке, она учитывает все верные строки и выводит найденный ответ.

```python
with open(r'task\calc.txt', 'r') as file:
    summa = 0
    for line in file:
        try:
            tmp = line.split()
            if tmp[1] == '+':
                summa += int(tmp[0]) + int(tmp[2])
            if tmp[1] == '-':
                summa += int(tmp[0]) - int(tmp[2])
            if tmp[1] == '*':
                summa += int(tmp[0]) * int(tmp[2])
            if tmp[1] == '/':
                summa += int(tmp[0]) / int(tmp[2])
            if tmp[1] == '%':
                summa += int(tmp[0]) % int(tmp[2])
            if tmp[1] == '//':
                summa += int(tmp[0]) // int(tmp[2])
        except ZeroDivisionError:
            print('На ноль делить нельзя!')

print('Сумма результатов:', summa)
```

```python
# Вариант 1:

user_name = input('Как вас зовут? ')
while True:
    print('Чтобы увидеть текущий текст чата введите 1, чтобы написать сообщение введите 2')
    response = input('Введите 1 или 2: ')
    if response == 'Стоп':   
        print('Завершение программы')
        break
    if response == '1':
        try:
            with open('chat.txt', 'r') as file:
                messages = file.readlines()
                print(''.join(messages))
        except FileNotFoundError:
            print('Служебное сообщение: пока ничего нет\n')
    elif response == '2':
        new_message = input('Введите сообщение: ')
        with open('chat.txt', 'a') as file:
            file.write('{name}: {message}\n'.format(name=user_name, message=new_message))
    else:
        print('Неизвестная команда\n')

# Основное преимущество программы в том, что здесь есть, куда разгуляться
# Например, можно сделать так, чтобы раз в какое-то кол-во секунд обновленный чат сам выводился на экран
# С этим может помочь модуль time 
# Или добавить команды для смайликов
# или вместо консоли сделать полноценную приятную оболочку
# А если изучить, как работают сети и различные протоколы, то полноценное приложение получится
```

```python
# Вариант 2:

from datetime import datetime

def identificftion():
    name = input('Введите имя: ')
    if name != '':
        return name
    else:
        identificftion()
        
def actions(nick):
    print('1. Посмотреть текущий текст чата')
    print('2. Отправить сообщение')
    try:
        selected_action = int(input('Выберите одно из действий: '))
        if selected_action == 1:
            print(read_history(chat_history))
            actions(nick)
        elif selected_action == 2:
            send_message(nick)
        else:
            actions(nick)
    except ValueError:
        print('Неверный выбор')
        actions(nick)
    except FileNotFoundError:
        print('История чата отсутсвует')
        actions(nick)
        
def send_message(nick):
    if (msg := input('msg: ')) != 'exit':
        save_history(chat_history, f'User: {nick} [{datetime.now()}]\nMessage: {msg}\n{"-" * 40}')
    actions(nick)
        
def save_history(filename, message):
    with open(filename, 'a') as file:
        file.write(message + '\n')
        
def read_history(filename):
    with open(filename, 'r') as file:
        lines = file.read()
    return lines

chat_history = 'history.txt'

nickname = identificftion()
print(f'Привет, {nickname}')
actions(nickname)
```
